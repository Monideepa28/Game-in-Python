```python
from sre_constants import REPEAT
import random
print("--Welcome to RPS--\n")
user_score =0
comp_score =0
ties=0
name=input("Enter your name here: ")
print("""
  WINNING RULES ::
  1. Paper vs Rock -->Paper
  2. Rock vs Scissors --> Rock
  3. Scissors vs Paper --> Scissors\n""")

print("""CHOICES ARE: 
  1. Rock
  2. Paper
  3. Scissors\n""")
while True:
  choice = int(input("Enter your choice from 1-3: "))
  while choice>3 or choice <1:
    choice = int(input("Enter valid input: "))

  if choice==1:
    user_choice="Rock"
  elif choice==2:
    user_choice="Paper" 
  else:
    user_choice="Scissors" 

  print("The user's choice is: ",user_choice)
  print("Now it is Computer's choice time..")
  computer =random.randint(1,3)
  if choice==1:
    computer_choice="Paper"
  elif choice==2:
    computer_choice="Rock" 
  else:
    computer_choice="Scissors" 

  print("The computer's choice is: ",computer_choice)
  if((user_choice == "Paper" and computer_choice == "Rock")or(user_choice == "Rock" and computer_choice == "Paper")):
    print("Paper wins")
    result= "Paper"
  elif((user_choice == "Scissors" and computer_choice == "Rock")or(user_choice == "Rock" and computer_choice == "Scissors")):
    print("Rock wins")
    result= "Rock"
  elif(user_choice == computer_choice):
    print("It's tie...")
    result = "Tie"
  else:
    print("Scissors wins")
    result ="Scissors"


  if result == "Tie":
    ties +=1
  elif result == user_choice:
    user_score +=1
  else:
    comp_score +=1

  print("\n---Score Board---")
  print(name,"'s score is ",user_score)
  print("Computer's score is ",comp_score)
  print("Ties are ",ties)

  REPEAT = input("Do you want to play again??")
  if REPEAT == "No" or REPEAT =="NO" or REPEAT =="no":
    break

print("\n Game Over!!")
print("\nTHANKS FOR PLAY.....")
```
